### Data Preprocessing

In this part of work, i preproccess the data. From data understanding we got that some original text becomes more toxic instead of be neutral. I decided to create new columns 'toxic' and 'neutral' to combine there original and translated text by comparing initial and final toxicity score to separate data in proper way. So, after that manupulation, the toxic text toxicity mean value become 0.94 and neutral text toxicity 0.03.

```python
import pandas as pd

df = pd.read_csv("../data/interim/data1.csv")

# Refactor toxic and neutral columns
df['toxic'] = df.apply(lambda row: row['translation'] if row['ref_tox'] < row['trn_tox'] else row['reference'], axis=1)
df['neutral'] = df.apply(lambda row: row['reference'] if row['ref_tox'] < row['trn_tox'] else row['translation'], axis=1)

df['toxicity'] = df.apply(lambda row: row['trn_tox'] if row['ref_tox'] < row['trn_tox'] else row['ref_tox'], axis=1)
df['neutral_toxicity'] = df.apply(lambda row: row['ref_tox'] if row['ref_tox'] < row['trn_tox'] else row['trn_tox'], axis=1)

# make all text lower-case
df['toxic'] = df['toxic'].str.lower()
df['neutral'] = df['neutral'].str.lower()

# drop old columns and lenght_diff
df.drop(['reference', 'translation', 'lenght_diff', 'ref_tox', 'trn_tox', 'index'], axis=1, inplace=True)
```
